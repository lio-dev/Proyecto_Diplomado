```python
import pandas as pd

df = pd.read_csv('preparacion de datos\Hoja 3_Resumen.csv', delimiter=';')
df
```

```python
import pandas as pd
import numpy as np

# Supongamos que tu DataFrame se llama 'df'
# Asegúrate de que la columna 'Año de Gestion' sea de tipo int para que la extrapolación funcione correctamente

# Verifica y convierte el tipo de dato si es necesario
df['Año de Gestion'] = df['Año de Gestion'].astype(int)

# Crea un DataFrame para almacenar la proyección
proyeccion_df = pd.DataFrame(columns=df.columns)

# Define los años para los que deseas hacer la proyección (2016 a 2025)
years_proyeccion = range(2016, 2031)

# Itera por cada departamento y realiza la extrapolación
for departamento in df['Departamento'].unique():
    # Filtra los datos por departamento
    departamento_df = df[df['Departamento'] == departamento]
    
    # Realiza la extrapolación utilizando numpy.polyfit
    coeficientes = np.polyfit(departamento_df['Año de Gestion'], departamento_df['Total Establecimientos De Salud'], deg=1)
    
    # Genera la proyección para los años deseados
    proyeccion_valores = np.polyval(coeficientes, years_proyeccion)
    
    # Redondea los valores a números enteros
    proyeccion_valores = np.round(proyeccion_valores).astype(int)
    
    # Crea un DataFrame temporal para almacenar la proyección del departamento actual
    proyeccion_departamento_df = pd.DataFrame({
        'Departamento': [departamento] * len(years_proyeccion),
        'Año de Gestion': years_proyeccion,
        'Total Establecimientos De Salud': proyeccion_valores
    })
    
    # Concatena el DataFrame temporal al DataFrame de proyección general
    proyeccion_df = pd.concat([proyeccion_df, proyeccion_departamento_df])

# Convierte la columna 'Total Establecimientos De Salud' a tipo entero
proyeccion_df['Total Establecimientos De Salud'] = proyeccion_df['Total Establecimientos De Salud'].astype(int)

# Muestra el DataFrame de proyección
print(proyeccion_df)
```
