```python
import pandas as pd

df_poblacion = pd.read_csv('preparacion de datos\indicadores_municipios_cpv-2012_cod.csv', delimiter=',')
df_poblacion
```

```python
import pandas as pd
import numpy as np

# Supongamos que tu DataFrame se llama 'df_poblacion'
# Asegúrate de que las columnas tengan los tipos de datos adecuados
df_poblacion['t_2001'] = df_poblacion['t_2001'].astype(int)
df_poblacion['t_2012'] = df_poblacion['t_2012'].astype(int)
df_poblacion['tac_ 2001_2012(%)'] = df_poblacion['tac_ 2001_2012(%)'].astype(float)

# Define los años para los que deseas hacer la proyección (2013 a 2030)
years_proyeccion = range(2013, 2031)

# Crea un DataFrame para almacenar la proyección de población
proyeccion_poblacion_df = pd.DataFrame()

# Itera por cada municipio y realiza la proyección
for index, row in df_poblacion.iterrows():
    poblacion_inicial = row['t_2012']
    tasa_crecimiento = row['tac_ 2001_2012(%)'] / 100
    
    # Realiza la proyección utilizando la fórmula del crecimiento exponencial
    proyeccion_poblacion = poblacion_inicial * (1 + tasa_crecimiento)**(np.array(years_proyeccion) - 2012)
    
    # Crea un DataFrame temporal para almacenar la proyección
    proyeccion_municipio_df = pd.DataFrame({
        'cod.mun': [row['cod.mun']] * len(years_proyeccion),
        'Año': years_proyeccion,
        'Poblacion Proyectada': proyeccion_poblacion.round().astype(int)
    })
    
    # Concatena el DataFrame temporal al DataFrame de proyección general
    proyeccion_poblacion_df = pd.concat([proyeccion_poblacion_df, proyeccion_municipio_df])

# Muestra o guarda el DataFrame de proyección de población
proyeccion_poblacion_df
# proyeccion_poblacion_df.to_csv('proyeccion_poblacion_municipios.csv', index=False)  # Si deseas guardar en un archivo CSV
```
