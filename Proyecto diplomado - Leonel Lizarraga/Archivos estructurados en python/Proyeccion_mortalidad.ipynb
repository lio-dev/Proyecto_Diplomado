```python
import pandas as pd
import numpy as np

# Supongamos que tu DataFrame se llama 'df_mortalidad'
# Asegúrate de que las columnas tengan los tipos de datos adecuados
df_mortalidad['Año de Gestion'] = df_mortalidad['Año de Gestion'].astype(int)

# Define los años para los que deseas hacer la proyección (2016 a 2030)
years_proyeccion = range(2016, 2031)

# Crea un DataFrame para almacenar la proyección de mortalidad
proyeccion_mortalidad_df = pd.DataFrame()

# Itera por cada municipio y departamento y realiza la proyección
for departamento, municipio in df_mortalidad[['Departamento', 'Municipio']].drop_duplicates().itertuples(index=False):
    # Filtra los datos por departamento y municipio
    dep_mun_df = df_mortalidad[(df_mortalidad['Departamento'] == departamento) & (df_mortalidad['Municipio'] == municipio)]
    
    # Crea un DataFrame temporal para almacenar la proyección
    proyeccion_mun_df = pd.DataFrame({
        'Departamento': [departamento] * len(years_proyeccion),
        'Municipio': [municipio] * len(years_proyeccion),
        'Año de Gestion': years_proyeccion
    })
    
    # Itera por cada tipo de mortalidad y realiza la proyección
    for tipo_mortalidad in ['mortalidad_fetal', 'Nacidos Muertos Atendidos Por Partera', 'Nacidos Muertos Atendidos Por Personal De Salud', 'Nacidos Muertos']:
        # Realiza la extrapolación utilizando numpy.polyfit
        coeficientes = np.polyfit(dep_mun_df['Año de Gestion'], dep_mun_df[tipo_mortalidad], deg=1)
        proyeccion_valores = np.polyval(coeficientes, years_proyeccion)
        
        # Asegúrate de que los valores no sean menores que cero
        proyeccion_valores = np.maximum(proyeccion_valores, 0)
        
        # Redondea los valores a números enteros
        proyeccion_valores = np.round(proyeccion_valores).astype(int)
        
        # Crea una columna en el DataFrame temporal para el tipo de mortalidad
        proyeccion_mun_df[tipo_mortalidad] = proyeccion_valores
    
    # Concatena el DataFrame temporal al DataFrame de proyección general
    proyeccion_mortalidad_df = pd.concat([proyeccion_mortalidad_df, proyeccion_mun_df])

# Muestra o guarda el DataFrame de proyección de mortalidad
proyeccion_mortalidad_df
# proyeccion_mortalidad_df.to_csv('proyeccion_mortalidad.csv', index=False)  # Si deseas guardar en un archivo CSV
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Carga de datos
df = pd.read_csv("preparacion de datos\mortalidad_fetal.csv", delimiter=';')

# Preprocesamiento de datos
df.fillna(0, inplace=True)

# Creación del modelo de regresión
X = df[['Año de Gestion']]  # Variable independiente (año)
y = df['mortalidad_fetal']   # Variable dependiente (mortalidad fetal)

model = LinearRegression()
model.fit(X, y)

# Proyección hasta 2030
future_years = np.arange(2016, 2031).reshape(-1, 1)
future_predictions = model.predict(future_years)

# Crear un DataFrame con las proyecciones
future_df = pd.DataFrame({'Año de Gestion': future_years.flatten(), 'Mortalidad Fetal Proyectada': future_predictions})

# Mostrar la tabla final
print(future_df)
```

```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
import numpy as np

# Carga de datos
df = pd.read_csv("preparacion de datos\mortalidad_fetal.csv", delimiter=';')


# Preprocesamiento de datos
df.fillna(0, inplace=True)

# Creación del modelo de regresión polinómica
X = df[['Año de Gestion']]  # Variable independiente (año)
y = df['mortalidad_fetal']   # Variable dependiente (mortalidad fetal)

degree = 2  # Grado del polinomio, puedes ajustarlo según lo que consideres apropiado

model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model.fit(X, y)

# Proyección hasta 2030
future_years = np.arange(2016, 2031).reshape(-1, 1)
future_predictions = model.predict(future_years)

# Crear un DataFrame con las proyecciones
future_df = pd.DataFrame({'Año de Gestion': future_years.flatten(), 'Mortalidad Fetal Proyectada': future_predictions})

# Mostrar la tabla final
print(future_df)
```
